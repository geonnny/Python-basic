### 슬라이싱(문자열)

- 0부터 시작함. 
- [a:b]는 [a,b) 를 뜻함
- [a:] 는 a부터 끝까지, [:b]는 처음부터 b-1까지.
- (-)이용해서 뒤에서부터 가져올 수 있음.
- 밑에 두개 안되는거 참고.
- 공백도 하나의 숫자로 침.

```python
num = "970215-1109719"

print("성별 : " + num[7])
print("연 : " + num[0:2]) # num[:2]
print("뒷자리: ",num[-7:])
print("뒷자리: ",num[-7:0])    # 안뜸
print("뒷자리: ",num[-1:-7])   # 안뜸

print("뒷자리: ",num[-7:-1])   # 다른게 뜸
```

### 문자열 처리 함수

- x.replace("abc","") 로 ""안에 아무것도 안넣음으로써 abc를 지울 수 있음.
- x.index("a") 는 '첫번째' a의 위치를 찾아줌.
- 한 문자가 아닌 여러 문자를 x.index에 넣으면 시작점을 반환함.(4~5인데 4반환)
- x.find()는 없는 것에 대해서 -1 반환. index는 오류뜸.

```python
sen = "Python is Amazing" 
print(sen.lower())        # 소문자로
print(sen.upper())        # 대문자로
print(sen[0].isupper())   # 선택된 문자가 대문자?
print(len(sen))           # 길이
print(sen.replace("Python", "Java")) # A를 B로 교체

idx = sen.index("n")      
print(idx)     
idx = sen.index("n", idx + 1)    # 두번째 n 찾기
print(idx)
idx = sen.index("on")
print(idx)


print(sen.find("Java"))
print(sen.find("on"))     # 이건 출력 결과 같음.

print(sen.count("n"))     # 몇개?
```

### 문자열 포맷

- %c 에 대해서 숫자를 넣으면 이상한 문자를 가져옴. 어디서 가져오는지는 모르겠음.
- s가 범용적으로 다 가능해서 s만 알고 있으면 될듯.
- format을 쓸거면 "".format임 . 뺴먹으면 안됨
- format에서 3번째 예시처럼 응용가능. s보다 format이 더 좋은듯.
- .format 안에서 저장된 변수들은 global 변수가 아님
- 방법3에서는 앞에 print문 내에서 f로 시작함으로써 실제 변수 저장된 값들을 가져옴.

```python
# 방법1
print("나는 %d살입니다." % 20)        # d 는 정수형만 가능,
print("나는 %d살입니다." % 20.1)      

print("나는 %s을 좋아해요." %"파이썬") # s는 str
print("Apple은 %c로 시작해요" % "A")    # 한 글자만 가져올 수있음.
#print("Apple은 %c로 시작해요" % "Ab")   오류뜸.
print("Apple은 %c로 시작해요" % 3)

print("나는 %s색과 %s색을 좋아해요." % ("파란","빨간"))  # 여러개
```

```python
# 방법2
print("나는 {}살입니다.". format("20"))
print("나는 {}색과 {}색을 좋아해요. ".format("파란","빨간")) 
print("나는 {1}색과 {0}색을 좋아해요. ".format("파란","빨간")) 

#print("나는 {}살이며, {}색을 좋아해요.".format(age=20,color="빨간")) 오류남
print("나는 {age}살이며, {color}색을 좋아해요.".format(color="빨간",age=20,))

# 방법3
age = 20
color = "빨간"
print(f"나는 {age}살이며, {color}색을 좋아해요.")
```

### 탈출문자(\관련)

- \n : 줄바꿈
- \" ,\' : 문장 내에서 " (')역할
- \\ : 문장 내에서 \
- \r : 커서를 맨 앞으로
- \b : 한 글자 삭제
- \t : 탭(큰 공백)
- 공백도 하나의 문자로 취급하여, 위\b 처럼 붙여써도 되고 아래처럼 띄워써도 됨.

```python
#print("바보
#입니다")   오류남
print("바보\n입니다")   # 줄바꿈

#print("메롱 "메롱"이다") 오류
print("메롱 '메롱' 이다")   # ""와 ''섞어쓰기. 반대도 됨.
print("메롱 \"메롱\"이다")  # \" : 문장 내에서 따옴표 가능. \'도 가능.

print("Red Apple\rPine")    # Red + 공백이 Pine으로 바뀜
print("Orange Apple\rPine")

print("Redd\bApple")
print("Redd \bApple")       # 여기선 공백 삭제

print("Red \t Apple")
```
