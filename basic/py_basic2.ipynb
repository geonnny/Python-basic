### 리스트

- .pop()은 마지막 사람 출력 후 list에서 제거
- 문자열 변수처럼 .index, count 다 가능. (똑같이 맨 앞은 0)
- x.reverse() 혹은 x.asd() 함수는 바로 적용을 하고 저장을 해버림. 출력은 x

```python
subway = [10, 20, 30]
print(subway)

subway = ["유재석", "조세호", "박명수"]
print(subway)

print(subway.index("조세호"))  # 조세호씨가 몇번째 칸에 ?

subway.insert(1, "정형돈")     # 1번째에 들어감. 조세호부터 한칸씩 밀림
print(subway)

print(subway.pop())            # 맨 마지막 사람을 출력하고 list에서 제거
print(subway)    

print(subway.pop())           
print(subway)    

subway.append("유재석")
print(subway.count("유재석"))
```

```python
num_list = [5,2,4,3,1]
num_list.sort()          # 정렬


print(num_list)

num_list.reverse()
print(num_list)

num_list.clear()    # 모두 지우기
print(num_list)


# 리스트 확장
num_list = [5,2,4,3,1]
mix_list = ["조세호",20,True]
num_list.extend(mix_list)
print(num_list)
```

### 사전

- key : value 순
- .get() 은 없는 key값을 넣었을때 None 출력 후 아래 코드를 실행
- 하지만 [] 는 없는 키 값을 넣었을 때 오류가 뜨며 그 이후 코드 실행 x
- .get(#,a) 는 # 이 있다면 출력 ,없으면 a를 출력
- key 값으로 str 도 사용 가능
- 추가할 때 기존에 있는 값이면 update를 함. 없으면 추가

```python
cabinet = {3 : "유재석", 100 : "김태호"}

print(cabinet[3])       # 방법1
print(cabinet.get(3))   # 방법2
print(cabinet.get(5, "available"))

# 키가 있는지 확인
print(3 in cabinet)
print(5 in cabinet)

# 추가
cabinet["C-20"] = "조세호"
print(cabinet)

# 제거
del cabinet["C-20"]
print(cabinet)

# key들만, value들만, 둘 다 출력
print(cabinet.keys())
print(cabinet.values())
print(cabinet.items())

# 모두 제거
cabinet.clear()
print(cabinet)
```

### 집합

- 중복 x, 순서 없음

```python
my_set = {1,2,3,3,3}
print(my_set)

java= {"a","b","c"}
python = set(["a","b"])    # 이런식으로도 가능

# 교집합
print(java&python)    
print(java.intersection(python))

# 합집합
print(java|python)
print(java.union(python))

# 차집합
print(java - python)
print(java.difference(python))

# 추가
python.add(1)
print(python)

# 제거
java.remove("a")
print(java)
```

### continue & break

- continue: 더 이상 아래 명령(이 예제에선 print)를 실행하지말고 다음번호로 넘어가라.

### 한 줄 For

```python
# 출석번호가 1,2,3,4 -> 100을 붙이기로함 -> 101,102,103,104

students = [1,2,3,4,5]
print(students, type(students))
students = [i + 100 for i in students]
print(students)

# 학생 이름을 길이로 변환
students = ["Iron man", "Thor","I an groot"]
students = [len(i) for i in students]
print(students)

# 학생 이름을 대문자로 변환
students = ["Iron man", "Thor","I an groot"]
students = [i.upper() for i in students]
print(students)
```
