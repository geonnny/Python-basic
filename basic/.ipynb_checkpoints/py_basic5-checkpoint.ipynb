## 예외처리

- try:, except: 이용
- 여러 에러가 있음 ex) ValueError, ZeroDivisionError

```python
print("나누기 전용 계산기입니다.")
num1 = int(input("첫 번째 숫자를 입력하세요 : "))
num2 = int(input("두 번째 숫자를 입력하세요 : "))
print("{0} / {1} = {2}".format(num1,num2, int(num1/num2)))

# 여기에 예를 들어서 3 말고 "삼" 같은것을 넣으면 실행안됨
```

```python
try:
    print("나누기 전용 계산기입니다.")
    num1 = int(input("첫 번째 숫자를 입력하세요 : "))
    num2 = int(input("두 번째 숫자를 입력하세요 : "))
    print("{0} / {1} = {2}".format(num1,num2, int(num1/num2)))
except ValueError:
    print("에러! 잘못된 값을 입력하였습니다.")
# 하지만 여기(위) 까지는 분모 0일때 오류가 뜸

except ZeroDivisionError as err:
    print(err)
    
```

```python
try:
    print("나누기 전용 계산기입니다.")
    nums = []
    nums.append(int(input("첫 번째 숫자를 입력하세요 : ")))
    nums.append(int(input("두 번째 숫자를 입력하세요 : ")))
    # 실수로 이 아래값을 리스트에 안넣었다고 가정
    # 즉 리스트 idx error
    #nums.append(int(nums[0] / nums[1]))
    print("{0} / {1} = {2}".format(nums[0],nums[1], nums[2]))
except ValueError:
    print("에러! 잘못된 값을 입력하였습니다.")

except ZeroDivisionError as err:
    print(err)
    
# 무슨 에러인지 알고 싶은 경우
except Exception as err:
    print("알 수 없는 에러가 발생하였습니다.")
    print(err)

# 무슨 에러인지 관심없고 나머지 모든 에러를 처리할 경우
#except:
#    print("알 수 없는 에러가 발생하였습니다.") 
```

## 에러 발생시키기

```python
try:
    print("한 자리 숫자 나누기 전용 계산기입니다.")
    num1 = int(input("첫 번째 숫자를 입력하세요 : "))
    num2 = int(input("두 번째 숫자를 입력하세요 : "))
    if num1 >= 10 or num2 >= 10:
        raise ValueError         # 의도적으로 에러 발생시키기 -> Except로 내려감.
    print("{0} / {1} = {2}".format(num1,num2, int(num1/num2)))
except ValueError:
    print("잘못된 값을 입력하였습니다. 한 자리 숫자만 입력하세요.")
```

## 사용자 정의 예외처리

- valueError, ZerodivisionError 같은 것은 파이썬에서 미리 제공하는 에러.
- 직접 정의해서 에러를 만들기 

```python
class BigNumberError(Exception):    # Exception 넣기.
    pass 
try:
    print("한 자리 숫자 나누기 전용 계산기입니다.")
    num1 = int(input("첫 번째 숫자를 입력하세요 : "))
    num2 = int(input("두 번째 숫자를 입력하세요 : "))
    if num1 >= 10 or num2 >= 10:
        raise BigNumberError
    print("{0} / {1} = {2}".format(num1,num2, int(num1/num2)))
except ValueError:
    print("잘못된 값을 입력하였습니다. 한 자리 숫자만 입력하세요.")
except BigNumberError:
    print("에러가 발생하였습니다.. 한 자리 숫자만 입력하세요.")
```

```python
# class 안에 메세지를 정의하고 싶을 경우

class BigNumberError(Exception):
    def __init__(self,msg):
        self.msg = msg
    
    def __str__(self):
        return self.msg
try:
    print("한 자리 숫자 나누기 전용 계산기입니다.")
    num1 = int(input("첫 번째 숫자를 입력하세요 : "))
    num2 = int(input("두 번째 숫자를 입력하세요 : "))
    if num1 >= 10 or num2 >= 10:
        raise BigNumberError("입력값: {0}, {1}".format(num1,num2,int(num1/num2)))
    print("{0} / {1} = {2}".format(num1,num2, int(num1/num2)))
except ValueError:
    print("잘못된 값을 입력하였습니다. 한 자리 숫자만 입력하세요.")
except BigNumberError as err:
    print("에러가 발생하였습니다.. 한 자리 숫자만 입력하세요.")
    print(err)
```

## Finally

- 예외처리 구문에서 정상적으로 수행되건, 오류가 발생하던 상관없이 맨 마지막에 실행되는 구문
- try 문에서 맨 마지막에 쓸 수 있음.

```python
# class 안에 메세지를 정의하고 싶을 경우

class BigNumberError(Exception):
    def __init__(self,msg):
        self.msg = msg
    
    def __str__(self):
        return self.msg
try:
    print("한 자리 숫자 나누기 전용 계산기입니다.")
    num1 = int(input("첫 번째 숫자를 입력하세요 : "))
    num2 = int(input("두 번째 숫자를 입력하세요 : "))
    if num1 >= 10 or num2 >= 10:
        raise BigNumberError("입력값: {0}, {1}".format(num1,num2,int(num1/num2)))
    print("{0} / {1} = {2}".format(num1,num2, int(num1/num2)))
except ValueError:
    print("잘못된 값을 입력하였습니다. 한 자리 숫자만 입력하세요.")
except BigNumberError as err:
    print("에러가 발생하였습니다.. 한 자리 숫자만 입력하세요.")
    print(err)
finally:
    print("계산기를 이용해 주셔서 감사합니다.")
```

## 외장함수

- 구글에 list of python modules 검색
- glob, os, time

```python
# glob
# 경로 내의 폴더 / 파일 목록 조회 (윈도우 dir)
import glob
print(glob.glob("*.ipynb"))

# os : 운영체제에서 제공하는 기본 기능
import os
print(os.getcwd())   # 현재 디렉토리 표기

folder = "sample_dir"
if os.path.exists(folder):   # 폴더가 (괄호 안에 있는게)있으면 실행 
    print("이미 존재하는 폴더입니다.")
    os.rmdir(folder)     # 폴더 지우기
    print(folder, "폴더를 삭제하였습니다.")
else:
    os.makedirs(folder) # 폴더 생성
    print(folder, "폴더를 생성하였습니다.")
    
print(os.listdir())    # glob과 비슷
```

```python
# time : 시간 관련 함수들
import time
print(time.localtime())
print(time.strftime("%Y-%m-%d %H:%M:%S"))

import datetime
print("오늘 날짜는 ", datetime.date.today())

# timedelta : 두 날짜 사이의 간격
today = datetime.date.today()
td = datetime.timedelta(days= 100)    # 100일을 저장
print("우리가 만난지 100일은", today + td)   # 오늘 날짜로부터 100일 뒤
```
