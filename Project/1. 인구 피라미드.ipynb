# 1. 인구 피라미드

## 2011년

## 남자 데이터 정의

```python
import pandas as pd
df_m = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx',skiprows=3, index_col='행정기관', 
                     usecols='B,E:Y')   # usecols : 원하는 열만 불러오기 B와 E부터 Y까지
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int) # 1,195,951(문자형) -> 1195951 (정수형)
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx',skiprows=3, index_col='행정기관', 
                     usecols='B,AB:AV')   # usecols : 원하는 열만 불러오기 B와 E부터 Y까지
df_w.head(3)
```

- .1 이라는거 붙는 이유 : 저 변수명이 2번째라서(남자에 똑같은 컬럼명 있어서)

```python
df_w.columns = df_m.columns  # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int) 
df_w.head(3)
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False  
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0]/ 1000)  # 단위 : 천 명
# 남자 데이터 앞에 -를 붙임으로써 대칭 형태로 보이게 함
plt.barh(df_w.columns, df_w.iloc[0]/ 1000)
plt.title('2011 대한민국 인구 피라미드 (단위 : 천 명)')
plt.savefig('2011_인구피라미드.png',dpi=100)
plt.show()
```
