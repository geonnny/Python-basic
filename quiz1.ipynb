## Quiz3)

예) http://naver.com  
규칙1) : http:// 부분은 제외 => naver.com  
규칙2) : 처음 만나는 점(.) 이후 부분은 제외 => naver  
규칙3) : 남은 글자 중 처음 세자리 + 글자 갯수 + 글자 내 'e'의 갯수 + "!" 로 구성

예) 생성된비밀번호 : nav51!


```python
lin = "http://naver.com"

loc1 = lin.find('/')
loc2 = lin.find('/', loc1+1)
loc3 = lin.find('.')

word = lin[loc2+1:loc3]
print(word[:3],len(word),word.count('e'),"!") # , 로 묶으면 중간에 공백이 생김
print(word[:3]+str(len(word))+str(word.count('e'))+"!")


# 풀이 방법
url = "http://naver.com"
my_str = url.replace("http://","")  # ""안에 아무것도 안넣음으로써 지움
my_str = my_str[:my_str.index(".")]  # . 앞까지 선택.
password = my_str[:3] + str(len(my_str)) + str(my_str.count("e")) + "!"
print(password)
```
