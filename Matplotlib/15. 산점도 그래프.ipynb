# 15. 산점도 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False  
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

### 점 크기 설정

```python
import numpy as np
sizes = np.random.random(8) * 1000  # (0 ~1 사이값) * 1000

plt.scatter(df['영어'],df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500  # 1,2,3 학년 각각 500,1000, 1500

plt.scatter(df['영어'],df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

### 색 삽입 : cmap

#### 참고

matplotlib cmap

https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
plt.scatter(df['영어'],df['수학'], s=sizes, c=df['학년'],  cmap = 'viridis', alpha = 0.7)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize = (7,7))
plt.scatter(df['영어'],df['수학'], s=sizes, c=df['학년'],  cmap = 'viridis', alpha = 0.7)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar()
```

```python
# 근데 위처럼 2.5 , 1.75 같은 값은 필요 없기에

plt.figure(figsize = (7,7))
plt.scatter(df['영어'],df['수학'], s=sizes, c=df['학년'],  cmap = 'viridis', alpha = 0.7)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks = [1,2,3] , label = '학년', shrink = 0.5,orientation = 'horizontal' ) 
# shrink = 컬러바 사이즈
# orientation = 'horizontal'  : 컬러바 아래쪽에 수평으로
```
