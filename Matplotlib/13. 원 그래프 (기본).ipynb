# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False  
```

```python
values = [30, 25, 20, 13, 10,2 ]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values)
plt.show()
```

```python
values = [30, 25, 20, 13, 10,2 ]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels = labels)
plt.show()
```

### 비율 넣기 : autopct

```python
values = [30, 25, 20, 13, 10,2 ]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct= '%.1f')    # %를 소수점 첫번째 자리까지
plt.show()
```

```python
values = [30, 25, 20, 13, 10,2 ]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct= '%.1f%%')    # %% 두개 더 붙이면 %까지 붙음
plt.show()
```

### 각도 돌리기 : startangle

```python
values = [30, 25, 20, 13, 10,2 ]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct= '%.1f%%', startangle=90) 
# 원래는 0도에서부터 시작해서 반시계 방향. 이제는 90도부터 시작
plt.show()
```

```python
values = [30, 25, 20, 13, 10,2 ]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct= '%.1f%%', startangle=90,
       counterclock=False)     # 반시게 말고 시계방향으로 돌리기
plt.show()
```

### 각 파이를 중심에서부터 띄우기. 강조의 효과 : explode

```python
values = [30, 25, 20, 13, 10,2 ]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
explode = [0.2, 0.1, 0 , 0 , 0 , 0]  # 각각마다 띄울 정도

plt.pie(values, labels = labels , explode = explode)
plt.show()
```

```python
values = [30, 25, 20, 13, 10,2 ]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
explode = [0.05] *6  # 0.05가 6개 있는 배열

plt.pie(values, labels = labels , explode = explode)
plt.show()
```

### 범례 삽입

```python
plt.pie(values, labels = labels, explode = explode)
plt.legend(loc = (1.2,0.3))
plt.show()
```

```python
plt.pie(values, labels = labels, explode = explode)
plt.title('언어')
plt.legend(loc = (1.2,0.3), title = '언어별 선호도')   # 범례 타이틀 설정
plt.show()
```
