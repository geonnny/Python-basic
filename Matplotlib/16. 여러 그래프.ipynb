# 16. 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False  
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2,2, figsize =(15,10))   # 2x2 4개의 그래프를 그리겠다는 것
fig.suptitle('여러 그래프 넣기')   # 전체의 제목
```

```python
fig, axs = plt.subplots(2,2, figsize =(15,10))   
fig.suptitle('여러 그래프 넣기')

axs[0,0].bar(df['이름'],df['국어'], label='국어점수')
axs[0,0].set_title('첫 번째 그래프')    # 첫번째 그래프의 제목
axs[0,0].legend()
axs[0,0].set(xlabel = '이름', ylabel = '점수')  # 첫 번째 그래프의 x,y축 이름
axs[0,0].set_facecolor('lightyellow')   # 첫번째 그래프의 전경색
axs[0,0].grid(ls = '--',lw = 0.5)

axs[0,1].plot(df['이름'],df['수학'], label='수학 점수')
axs[0,1].plot(df['이름'],df['영어'], label='영어 점수')
axs[0,1].set_title('두 번째 그래프') 
axs[0,1].legend()


axs[1,0].barh(df['이름'], df['키'])

axs[1,1].plot(df['이름'], df['사회'])
```
