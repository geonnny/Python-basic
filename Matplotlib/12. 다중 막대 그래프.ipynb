```python
# plt.bar(x -1 ,y ) : x보다 약간 왼쪽에 그려줌
index = np.arange(df.shape[0])   # df 의 row 숫자만큼 array 만들어줌
index
```

```python
w = 0.25
plt.bar(index - w , df['국어'], width = w)   
plt.bar(index , df['영어'], width = w)
plt.bar(index + w , df['수학'], width = w)   
```

```python
plt.figure(figsize = (10,5))
plt.title('학생별 성적')

w = 0.25
plt.bar(index - w , df['국어'], width = w, label = '국어')   
plt.bar(index , df['영어'], width = w, label = '영어')
plt.bar(index + w , df['수학'], width = w, label = '수학')   
plt.legend(ncol = 3)
plt.xticks(index , rotation = 60)
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.title('학생별 성적')

w = 0.25
plt.bar(index - w , df['국어'], width = w, label = '국어')   
plt.bar(index , df['영어'], width = w, label = '영어')
plt.bar(index + w , df['수학'], width = w, label = '수학')   
plt.legend(ncol = 3)
plt.xticks(index , df['이름'], rotation = 60)  # index 값에 이름을 넣기
plt.show()
```
