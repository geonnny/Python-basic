## 집계 함수 (Aggregate Functions)
: 기본적으로 수에 대해서만 가능. NaN 뺴고 하려면 nan을 붙여야함

### np.sum() : 합 계산 (NaN 제외 : np.nansum())

```python
a2 = np.random.randint(1,10,size = (3,3))
print(a2)
print(a2.sum(), np.sum(a2))    # 전체합. 둘 다 가능
print(a2.sum(axis = 0), np.sum(a2, axis = 0))   # colsum
print(a2.sum(axis = 1), np.sum(a2, axis = 1))   # rowsum
```

### cumsum() : 누적합 계산 ( NaN 제외 : np.nancumsum())

```python
print(a2)
print(a2.cumsum(), np.cumsum(a2))    # 전체합. 둘 다 가능
print(np.cumsum(a2, axis = 0))
print(np.cumsum(a2, axis = 1))
```

### diff() : 차분 계산 (NaN 제외 : 존재x)

```python
print(a2)
print(np.diff(a2))    # [2열 - 1열 , 3열 - 1열]
print(np.diff(a2, axis = 0))  # [2행 - 1행, 3행 - 2행]
print(np.diff(a2, axis = 1))  # [2열 - 1열 , 3열 - 1열]. none 값과 같음
```

### prod() : 곱 계산 ( NaN 제외 : np. nanprod())

```python
print(a2)
print(a2.prod(), np.prod(a2))    #  둘 다 가능
print(np.prod(a2, axis = 0))
print(np.prod(a2, axis = 1))
```

### cumprod() : 누적 곱 계산

```python
print(a2)
print(a2.cumprod(), np.cumprod(a2))    #  둘 다 가능
print(np.cumprod(a2, axis = 0))
print(np.cumprod(a2, axis = 1))
```

### dot() / matnul : 점곱 / 행렬 곱 ( NaN 제외 : 존재 X)

```python
print(a2)
b2 = np.ones_like(a2)
print(b2)
print(np.dot(a2,b2))     # [a1행 x b1 행 전부 더한값이 1행]
print(np.matmul(a2,b2))  # 행렬 곱
```

### tensordot() : 텐서곱 계산 ( NaN 제외 : 존재 X)

```python
print(a2)
print(b2)
print(np.tensordot(a2,b2))              # 37. dot() 한거 전부 더한것
```

```python
c2 = np.arange(1,10).reshape(3,3)
print(np.tensordot(a2,c2,axes = 0))    # axis 아니고 axes
# a2 의 (1,1) 원소를 b2에 곱한것이 첫번쨰 층. a2의 (2,1) 를 b2에 곱한게 두번째 층
```

```python
# 전부 행렬 곱
print(np.tensordot(a2,c2, axes = 1))
print(np.dot(a2,c2))
print(np.matmul(a2,c2))
```

### cross() : 벡터 곱 ( NaN 제외 : 존재 X)

- 살짝 여인수 계산 느낌

```python
x = [1,2,3]
y = [4,5,6]

print(np.cross(x,y))
# [x[2]*y[3] - x[3]*y[2], x[3]*y[1] - x[1]*y[3] , x[3]*y[2] - x[2]*y[6]] 
```

```python
print(x*3) # 참고 리스트에 * 하면 그만큼 반복
#x +3 이건 안됨

x2 = np.array([1,2,3])
y2 = np.array([4,5,6])
np.cross(x2,y2)       # 리스트, array 둘다 됨
```

## 비교 연산(Comparison Operators)

|연산자 | 비교 범용함수 | 
|:-----|:----:|
|``==`` |`` np.equal``|
|``!=`` |`` np.not_equal``|
|``<``| ``np.less``|
|``<=``| ``np.less_equal``|
|``>``| ``np.greater``|
|``>=``| ``np.greater_equal``|

```python
a1 = np.arange(1,10)
print(a1)
print(a1 == 5)
print(np.equal(a1,5))  # 위와 같음
```

```python
b1 = np.arange(1,4)
print(b1)
a1==b1    # size 다르면 안됨
```

```python
a2 = np.arange(1,10).reshape(3,3)
print(a2)
print(np.equal(a2,5)) # 2차원도 가능
```

```python
print(np.count_nonzero(a2>5))  # 조건 붙이기 가능
print(np.sum(a2>5))            # 위와 같음
print(np.sum(a2>5, axis = 0))
print(np.sum(a2>5, axis = 1))
print(np.any(a2>5, axis = 1))
```

|비교 범용 함수 | 설명 |
|:-----|:----:|
| ``np.isclose``| 배열 두개가 (z*1e+02)% 내외로 가까우면 True, 아니면 False | 
| ``np.isinf``|배열이 inf이면 True, 아니면 False| 
| ``np.isfinite``| 배열이 inf,nan이면 False, 아니면 True|

```python
a1 = np.array([np.nan, 2, np.inf, 4, np.NINF])   # NINF : negative Inf
print(a1)
print(np.isnan(a1))
print(np.isinf(a1))
print(np.isfinite(a1))
```

### 불리언 연산자(Boolean Operators)

|연산자 | 비교 범용 함수 |
|:-----|:----:|
|``&`` | ``np.bitwise_and``|
|``\`` | ``np.bitwise_or`` |
|``^`` | ``np.bitwise_xor``|
|``~`` | ``np.bitwise_not``|

- 위에서 사실은 \ 아니고 |
- xor 은 or을 제외한 것

```python
a2 = np.arange(1,10).reshape(3,3)
print(a2)
print((a2>5) & (a2 < 8))
print(a2[(a2>5) & (a2 < 8)])
```

```python
print((a2>5) | (a2 < 8))
print(a2[(a2>5) | (a2 < 8)])
```

```python
print((a2>5) ^ (a2 < 8))
print(a2[(a2>5) ^ (a2 < 8)])
```

```python
print(~(a2>5) )
print(a2[~(a2>5)])
```

### 부분 정렬

- partition() : 배열에서 k개의 작은 값을 반환

```python
a1 = np.random.randint(1,10,size = 10)
print(a1)
print(np.partition(a1,3))     # 3 개 값만을 앞으로 가져옴. 그 3개는 정렬 x
# 2, 3, 3 순서인데 그렇게 정렬 안되는거 보면
# 메모리 가장 최소화 하는 방향으로 정렬하는듯 ??
# 정렬 안된 나머지도 자기 멋대로 순서옴
```

```python
a2 = np.random.randint(1,10,size =(4,4))
print(a2)
print(np.partition(a2,2))    # 디폴트는 axis = 1
print(np.partition(a2,2, axis = 0))  # 정렬도 column 에서 함
print(np.partition(a2,2, axis = 1))
```
