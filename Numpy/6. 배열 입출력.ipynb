# 6. 배열 입출력

```python
import numpy as np
```

|함수 | 설명 | 파일 종류|
|:-----|:----:|:----:|
| ``np.save()``|Numpy 배열 객체 1개를 파일에 저장 | 바이너리|
| ``np.savez()``|Numpy 배열 객체 여러개를 파일에 저장 | 바이너리 |
| ``np.load() ``|Numpy 배열 저장 파일로부터 객체 로딩| 바이너리|
| ``np.loadtxt() ``|텍스트 파일로부터 배열 로딩| 텍스트 |
| ``np.savetxt() ``|텍스트 파일에 Numpy 배열 객체 저장| 텍스트  |

```python
a2 = np.random.randint(1,10,size = (5,5))
print(a2)
np.save("a222", a2)
```

```python
%ls    # a222라는 npy파일이 생김
```

```python
b2 = np.random.randint(1,10,size = (5,5))
print(b2)
np.savez("ab222",a2,b2)
```

```python
%ls        # 여러개라서 ab222.npz 라는 파일이 생김
```

```python
npy = np.load("a222.npy")
print(npy)
```

```python
npz = np.load("ab222.npz")
print(npz.files)  # 여러개라서 .files
```

```python
print(npz['arr_0'])     # a2
print(npz['arr_1'])     # b2
```

```python
np.savetxt("a222.csv",a2, delimiter=',')   
# csv로 저장하고 구분자(delimiter)를 , 로 지정해주기 
```

```python
!more a222.csv
```

- 실수 형태를 텍스트 형태로 저장을 하기에 파일 열어보면 .0000000000 이런게 붙음 

```python
# 옵션
np.savetxt("b222.csv",a2, delimiter=',', fmt = '%.1e',   # 소수점 1개만
          header = 'c1,c2,c3,c4,c5') 
```

```python
!more b222.csv  
```

- header도 생기고, 소수점도 한자리까지만 생김

```python
np.savetxt("a222.txt",a2)
```
