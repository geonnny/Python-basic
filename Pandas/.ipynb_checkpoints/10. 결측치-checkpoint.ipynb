```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## 데이터 채우기 fillna

```python
df.fillna('')   # NaN 데이터를 빈칸으로 채움. 하지만 저장x 저장할거면 inplace=True
```

- fillna()는 전체 데이터에 대해서 적용

```python
import numpy as np
df['학교'] = np.nan      # 학교 데이터 전체를 NaN 으로 채움
df
```

```python
df.fillna('모름')
```

```python
df.fillna('모름', inplace=True)
df
```

```python
# 다시 원본 데이터 불러오기
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
# 특정 Column에 대해서만 적용하기
df['SW특기'].fillna('확인 중', inplace=True)
df
```

## 데이터 제외하기 dropna

```python
# 다시 원본 데이터 불러오기
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna(inplace=True)     # 전체 데이터 중에서 NaN을 포함하는 데이터 삭제
df
```

```python
# 다시 원본 데이터 불러오기
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

- axis : index or columns (index : Nan 포함 row 삭제, columns : 열 삭제)  
- how : any or all ( any : NaN 하나라도 있으면)  
- default : row, any

```python
df.dropna(axis = 'index', how = 'any')
```

```python
# NaN 이 하나라도 있는 row 삭제
df.dropna(axis = 'columns', how = 'any')
```

```python
# 학교 데이터를 모두 NaN 으로 변경
df['학교'] = np.nan
df
```

```python
# 데이터 전체가 NaN 인 경우에만 Column 삭제
df.dropna(axis = 'columns', how ='all')
```
