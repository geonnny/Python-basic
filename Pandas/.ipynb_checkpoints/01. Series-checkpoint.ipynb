# Pandas
: 파이썬에서 사용하는 데이터분석 라이브러리

```python
import pandas as pd
```

# 1. Series
1차원 데이터(정수, 실수, 문자열 등)

## Series 객체 생성
예) 1 ~ 4월 까지 평균 온도 데이터 (-20, -10, 10, 20)

```python
temp = pd.Series([-20, - 10, 10, 20])
temp
```

- 왼쪽은 idx를 나타냄

```python
temp[0] # 0번째 idx 출력. 1월온도
```

```python
temp[2] # 3월 온도
```

## Series 객체 생성 (Index 지정)

```python
temp = pd.Series([-20, - 10, 10, 20], index = ['Jan', 'Feb', 'Mar', 'Apr'])
temp
```

- 인덱스 부분이 바뀌었음

```python
temp['Jan']    # idx Jan 에 해당하는 데이터 출력
temp['Apr']  
```
