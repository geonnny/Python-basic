# 9. 데이터 선택 (조건)
: 조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185      # True / False 로 불러옴
```

```python
filt = (df['키'] >= 185  )
df[filt]
```

```python
df[~filt]     # ~ => False 데이터 불러옴
```

```python
df.loc[df['키'] >= 185  , ['수학','과학']]    # 키 185 이상 학생들의 수학,과학 점수
```

## 다양한 조건

### And 조건. (&)

```python
df.loc[(df['키'] >= 185)  & (df['학교']== '북산고')]
```

### Or 조건 (|)

```python
df.loc[(df['키'] < 170  ) | (df['키'] > 200  ), ['이름','수학']]
# 키가 170 미만이거나 200 초과인 사람들의 이름 , 수학 정보
```

### str 함수

```python
filt = df['이름'].str.startswith('송')  # '송'씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태')    # 이름에 '태'가 들어가는 사람
df[filt]
```

```python
df[~filt]      # 이름에 '태'가 들어가지 않는 사람
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)   # SW 특기가 Python 이거나 Java 인 사람
df[filt]
```

- 황태산은 PYTHON 이지만 대문자라서 안불러와짐
- 이런 경우엔 아래와 같이 소문자로 만들어서 비교하고 가져오기

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)   # .str.lower()을통해 소문자로 변환
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')    # Java 포함하는 것들 뽑기
df[filt]
```

```python
df['SW특기'].str.contains('Java')
```

- NaN 데이터 떄문에 이렇게 하면 오류 발생

```python
df['SW특기'].str.contains('Java' , na=False) # NaN 데이터에 대해서 False로 설정(True도 됨)
```

```python
filt = df['SW특기'].str.contains('Java', na = False)   # Java 포함하는 것들 뽑기
df[filt]
```
