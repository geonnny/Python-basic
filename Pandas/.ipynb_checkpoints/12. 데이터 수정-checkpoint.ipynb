# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
# Dictionary 형식으로
df['학교'].replace({'북산고' : '상북고'})   # 북산고 -> 상북고
```

```python
# 두 개 이상에 대해서 수정
df['학교'].replace({'북산고' : '상북고', '능남고': '무슨고'})
```

```python
df    # 역시 변화는 x inplace 넣어야함
```

```python
df['SW특기'].str.lower()
```

```python
# SW특기를 소문자로 저장
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
# SW특기를 대문자로 저장
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교'   # 학교 데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail'   # 결과 Column을 추가하고 전체 데이터는 Fail로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass'  # 총합 400 이상 데이터 Pass로
df
```

## Column 삭제

```python
df.drop(columns = ['총합'])    # 총합 Column 삭제. 역시나 반영 x
# 하나만 삭제할때는 [] 없애고 해도 됨
```

```python
# 여러개 삭제
df.drop(columns = ['총합', '영어'])  
```

## Row 삭제

```python
df.drop(index = '4번')   # 4번 학생 데이터 row를 삭제
```

## Column 순서 변경

```python
df.columns
```

```python
type(df.columns)
```

```python
# Column 이름을 리스트 형태로 가져옴
cols = list(df.columns)
cols
```

```python
type(cols)
```

```python
# 맨 뒤 결과 Column을 앞으로 가져오고 나머지 들과 합쳐서 순서 변경
df = df[[cols[-1]] + cols[0:-1]]
df
```

- cols[-1] 앞에만 [] 하나 더 붙는 이유 : 하나라서 list가 아닌데 list로 가져오려고
- [0:-1] -> 첫번째부터 맨 마지막 하나 전까지
