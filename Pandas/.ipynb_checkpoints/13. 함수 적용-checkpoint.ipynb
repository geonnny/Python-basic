- 키는 정수형 데이터인데 문자인 cm을 붙이려니까 오류 발생

## 데이터에 함수 적용 (apply)

```python
# Cm 을 붙여주는 함수 작성
def add_cm(height):
    return str(height) + 'cm'
```

```python
df['키'] = df['키'].apply(add_cm)
df
```

```python
df['키'].describe()  # 하지만 위에서 str 으로 바꿔서 문자 타입인 모습
```

```python
# 첫먼째 글자를 대문자로 나머지는 소문자로 바꿔주는 함수 작성
def capitalize(lang):
    if pd.notnull(lang):   # NaN인지 아닌지 
        return lang.capitalize()   # 여기서 capitalize 함수는 원래 있는 함수
    return lang
```

```python
df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

```python
df['SW특기'].str.capitalize()
# 원래는 str 함수에서 똑같은 기능을 해주는 함수가 있음
# 예시를 위해서 해본 것
```
