# 14. 그룹화
: 동일한 값을 가진 것들끼리 합쳐서 기초통계량 계산 위해 사용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.groupby('학교')
```

```python
# 그룹을 짓고 북산고 그룹만 가져오기
df.groupby('학교').get_group('북산고')
```

```python
df.groupby('학교').mean()   # 계산 가능한 데이터들의 평균값
```

```python
df.groupby('학교').size() # 각 그룹의 크기 정보
```

```python
df.groupby('학교').size()['능남고']   # 학교 그룹화 이후 능남고에 해당하는 데이터의 수
```

```python
df.groupby('학교')['키'].mean()   # 학교 그룹화 이후 키의 평균 데이터
```

- 어떤 특정 obs 불러올때는 맨 마지막에 붙고, 어떤 특정 변수에 대해 알아볼떄는 함수 전에

```python
df.groupby('학교')[['국어','영어','수학']].mean() 
# 여러개니까 [] 안에 한번 더 []로 lIst로 !
```

```python
# 학년이라는 새 변수 추가
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
df['학년'].dtypes    # 숫자형
```

### 여러 변수에 대해 그룹화

```python
df.groupby(['학교', '학년']).mean()  # 학교별,학년별 평균 데이터
```

- summary 같은거없나 ? 저런 요약에 각 셀 해당 데이터 수도 같이 나오는

```python
df.groupby('학년').mean()
```

```python
df.groupby('학년').mean().sort_values('키')
# 학년별로 키 평균값 내고 큰 키 순서로 오름차순 정렬
```

```python
df.groupby(['학교','학년']).sum()
```

```python
df.groupby('학교')[['이름','SW특기']].count()    
# 학교로 그룹화 한 뒤 학교별 SW특기 데이터의 수
# 북산고 두 명은 SW특기가 NaN 이라 카운트에서 제외됨
```

- size() 와는 또 다름.

```python
school = df.groupby('학교')
school['학년'].value_counts()   # 학교 그룹화 이후 학년별 학생 수
```

```python
school['학년'].value_counts().loc['북산고']   # 위 정보의 북산고만
```

```python
# 비율. normalize = True
school['학년'].value_counts(normalize = True).loc['북산고']
# 북산고 학생들의 학년 비율로 가져옴
```
