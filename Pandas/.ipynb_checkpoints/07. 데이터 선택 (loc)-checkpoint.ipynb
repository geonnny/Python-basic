# 7. 데이터 선택 (loc)
: 이름을 이용하여 원하는 row 에서 원하는 col 선택 (location)  
 -> Row 기준 !

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.loc['1번']
```

```python
df.loc['5번']
```

```python
df.loc['1번','국어']        # 1번에 해당하는 사람의 국어
```

```python
df.loc[['1번', '2번'],['국어', '영어']]     # 2명
```

```python
df.loc['1번':'5번', '국어':'사회']   # idx 1번부터 5번까지 (5번포함), 국어부터 사회까지 (사회 포함)
```

```python
df.loc['5번':'1번', '국어':'사회'] 
```

- row 순서 반대로 (위처럼) 입력하면 안나옴
