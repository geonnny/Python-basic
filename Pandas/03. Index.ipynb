```python
import pandas as pd

data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}


df = pd.DataFrame(data, index = ['1번','2번','3번', '4번','5번','6번','7번','8번'])
df
```

```python
df.index
```

## Index 이름 설정

```python
df.index.name = '지원번호'
df
```

## Index 초기화

- 실제 데이터에 반영하려면 inplace =True 넣기
- 칼럼으로 안받고 버릴려면 drop=True

```python
df.reset_index()
# - 0~7 이라는 새로운 idx가 생기고 기존의 1번 ~8번은 컬럼
```

```python
df.reset_index
# index() 안하면 이상한 형태로 출력 (보기 안좋음)
```

```python
df.reset_index(drop=True)
# 이렇게하면 col 으로 안받고 버림
```

```python
df
```

- 하지만 실제 데이터에는 반영이 안됨!

```python
df.reset_index(drop=True, inplace=True) # 실제데이터에 바로 반영
```

```python
df
```

## Index 설정
: 지정한 column으로 index를 설정

```python
df.set_index('이름')    # 이또한 실제 데이터에는 반영x
```

```python
df
```

```python
df.set_index('이름' , inplace=True)
```

## Index 정렬
: idx를 기준으로 오름차순, 내림차순 정렬

- 내림차순 : ascending = False

```python
df.sort_index()   # idx기준 오름차순 정렬. 역시나 반영은 x
```

```python
df
```

```python
df.sort_index(ascending=False)
```
