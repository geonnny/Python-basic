# 14. 그룹화
: 동일한 값을 가진 것들끼리 합쳐서 기초통계량 계산 위해 사용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.groupby('학교')
```

```python
# 그룹을 짓고 북산고 그룹만 가져오기
df.groupby('학교').get_group('북산고')
```

```python
df.groupby('학교').mean()   # 계산 가능한 데이터들의 평균값
```

```python
df.groupby('학교').size() # 각 그룹의 크기 정보
```

```python
df.groupby('학교').size()['능남고']   # 학교 그룹화 이후 능남고에 해당하는 데이터의 수
```

```python
df.groupby('학교')['키'].mean()   # 학교 그룹화 이후 키의 평균 데이터
```

```python
df.groupby('학교')[['국어','영어','수학']].mean() 
```

```python
# 학년이라는 새 변수 추가
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
df['학년'].describe()    # 숫자형
```

### 여러 변수에 대해 그룹화

```python
df.groupby(['학교', '학년']).mean()  # 학교별,학년별 평균 데이터
```

- summary 같은거없나 ? 저런 요약에 각 셀 해당 데이터 수도 같이 나오는

```python
df.groupby('학년').mean()
```
