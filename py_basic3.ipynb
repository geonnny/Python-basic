## 함수

```python
def open_account():
    print("새로운 계좌가 생성되었습니다.")
    
def deposit(balance,money):
    print("입금이 완료되었습니다. 잔액은{0} 원입니다.".format(balance + money))
    return balance + money    # 반환값 지정
```

```python
balance = 0
balance = deposit(balance, 1000)
print(balance)
```

```python
def withdraw(balance, money):
    if balance >= money:
        print("출금이 완료되었습니다. 잔액은 {} 원입니다.".format(balance - money))
        return balance - money
    else:
        print("출금 불가. 잔액은 {} 원입니다.".format(balance))
        return balance
    
balance = 0
balance = deposit(balance, 1000)
balance = withdraw(balance, 2000)
balance = withdraw(balance, 200)
```

```python
def withdraw_night(balance, money):
    commission = 100
    return commission, balance - money - commission # return 값이 튜플 형태로

commission , balance = withdraw_night(balance,500)
print("수수료 {} 원이며, 잔액은 {} 원입니다.".format(commission, balance))
```

## 가변인자

- 서로 다른 갯수의 값을 넣어줄 때 *을 활용해 가변인자를 쓸 수 있음.

```python
def profile(name, age, lang1, lang2, lang3, lang4, lang5):
    print("이름 :{0}\t나이 : {1}\t".format(name,age), end = " ") 
    # end 가 없으면 다음 프린트문을 줄바꿈해서 출력하지만 줄바꿈 하지 않고 출력
    print(lang1, lang2, lang3, lang4, lang5)
    
profile("유재석", 20, "Python", "Java", "C", "C++", "C#")
profile("김태호", 25, "Kotlin", "Swift", "","","")
```

```python
def profile(name, age, *language):    # *: 넣고싶은만큼 넣을수 있음.
    print("이름 :{0}\t나이 : {1}\t".format(name,age), end = " ") 
    # end 가 없으면 다음 프린트문을 줄바꿈해서 출력하지만 줄바꿈 하지 않고 출력
    for lang in language:
        print(lang, end=" ")
    print()
```

```python
profile("유재석", 20, "Python", "Java", "C", "C++", "C#", "Java")
profile("김태호", 25, "Kotlin", "Swift")
```

## 지역변수&전역변수

```python
# 오류 예시
gun = 10
def checkpoint(soldiers): # 근무 나가는 군인의 수
    gun = gun - soldiers            # 지역변수임. 여기서 gun은 밖 gun과 다름.
    # gun = 10 이렇게 지정을 해줘야함
    print("[함수 내] 남은 총 : {}".format(gun))
    
print("전체 총 : {0}".format(gun))
checkpoint(2)
print("남은 총 : {0}".format(gun))
```

```python
gun = 10
def checkpoint(soldiers): # 근무 나가는 군인의 수
    global gun # 전역 공간에 있는 gun 사용.
    gun = gun - soldiers            # 지역변수임. 여기서 gun은 밖 gun과 다름.
    print("[함수 내] 남은 총 : {}".format(gun))
    
print("전체 총 : {0}".format(gun))
checkpoint(2)
print("남은 총 : {0}".format(gun))
```

**위 방법은 전역변수가 많아지면 코드 관리 힘들어져서 권장 x**

**아래처럼 가급적 반환값을 받아서 사용하기**

```python
gun = 10
def checkpoint_ret(gun, soldiers):
    gun = gun - soldiers
    print("[함수 내] 남은 총 : {}".format(gun))
    return gun

print("전체 총 : {0}".format(gun))
gun = checkpoint_ret(gun, 2)
print("남은 총 : {0}".format(gun))
```

## 표준 입출력

- sep ( in print) :  각 각의 사이에 어떤 걸 넣을지 지정
- .l(r)just(x) : x만큼의 공간을 확보하고 왼(오른)쪽 정렬  
  -> 빈 공간이 자리수에 따라 다름.
- .zfill(x) : x만큼의 자리수를 확보하고 나머지 0 채우기

```python
print("Pyhton", "Java") # 중간에 띄어쓰기 있음
print("Pyhton"+ "Java") # 띄어쓰기 없음

print("Pyhton", "Java", sep = ",")   # 공백 대신 , 로 나옴. (default = 공백)
print("Pyhton", "Java", "Javascript", sep = " vs ")
```

```python
# 두 줄로 나옴
print("Pyhton", "Java", sep=",")
print("무엇이 더 재밌을까요?")

# end로 인해 한줄로 나옴
print("Pyhton", "Java", sep=",", end="?") # end = 끝부분을 ""로 바꿔달라
print("무엇이 더 재밌을까요?")
```

```python
import sys
print("Pyhton", "Java", file=sys.stdout)  # 표준 출력으로
print("Pyhton", "Java", file=sys.stderr)  # 표준 에러로
```

## 다양한 출력 포맷

```python
# 빈 자리는 빈공간으로 두고, 오른쪽 정렬을 하되, 총 10자리 공간을 확보
print("{0: >10}".format(500))  # : , > 사이에 공백있음
print("{0:>10}".format(500))  # : , > 사이에 공백없음. 위랑 똑같은디 ?


# 양수일 떈 +, 음수일 땐 - 로 표시
print("{0: >+10}".format(500))  # : , > 사이에 공백있음
print("{0: >+10}".format(-500))  # : , > 사이에 공백있음

# 왼쪽 정렬하고, 빈칸을 _로 채움
print("{0:_<+10}".format(500))  # 아하 공백이 default네
print("{0:_<10}".format(500))  # + 없어짐으로써 10자리중 하나 줄어서 _ 하나더

# 3자리 마다 콤마를 찍어주기
print("{0:,}".format(100000000000))

# 3자리 마다 콤마에 부호까지
print("{0:+,}".format(-100000000000))

# 3자리 마다 콤마, 부호, 자리수(30)까지 확보, 빈자리는 ^ 로 채우기
print("{0:^<+30,}".format(100000000000))
# print("{0:^+<30,}".format(100000000000)) 순서 중요한듯. 안돌아감


# 소수점 출력
print("{0:f}".format(5/3))

# 소수점 특정 자리수 까지만 출력( 3째 자리에서 반올림해서 2째 자리까지 표시)
print("{0:.2f}".format(5/3))
```

## 파일 입출력

- 첫번쨰 코드 돌리면 scre란 파일이 생김
- "w" : 쓰기. 다시 열면 덮어쓰기
- "a" : 이어쓰기 (append)
- "r" : 읽기
- .write() 는 자동으로 줄바꿈이 안됨.
- 파일 열면 닫는것까지 해야함

```python
score_file = open("score.txt", "w", encoding="utf8")
print("수학 : 0", file = score_file)
print("영어: 50", file = score_file)
score_file.close()
```

```python
score_file = open("score.txt", "a", encoding="utf8")
score_file.write("과학 : 80")
score_file.write("\n코딩 : 100") # 자동 줄바꿈 안돼서 줄바꾸기 추가 (따옴표 안에!)
score_file.close()
```

```python
score_file = open("score.txt", "r", encoding="utf8")
print(score_file.read())    # 한번에 전부 읽기
score_file.close()
```

```python
score_file = open("score.txt", "r", encoding="utf8")
print(score_file.readline())    # 줄별로 읽기, 읽고 커서는 다음줄로 이동
print(score_file.readline(),end="")    # 여기선 자동으로 줄바꿈 있음
print(score_file.readline())    # 줄바꿈 안하고 싶으면 위처럼 end 추가
score_file.close()
```

```python
# 여기선 우리가 4줄밖에 없는거 알지만 몇줄이 있는지 모르는 경우
score_file = open("score.txt", "r", encoding="utf8")
while True: # 무한루프
    line = score_file.readline()
    if not line:     # 읽어올 내용이 없으면
        break
    print(line, end="")
score_file.close()
```

```python
# 리스트에 다 넣어서 처리하기
score_file = open("score.txt", "r", encoding="utf8")
lines = score_file.readlines()  #  모든 line을 list 형태로 저장
for line in lines:
    print(line, end="")
    
score_file.close()
```

## pickle

- 프로그램 상에서 사용하고 있는 데이터를 파일 형태로 저장해주는 것.
- wb : w + b. 쓰기인데 binary 타입. pickle 위해선 binary 타입 정의 해줘야함.
- pickle에선 따로 encoding 지정 필요 x

```python
import pickle
profile_file = open("profile.pickle", "wb")
profile = {"이름" : "박명수", "나이":30, "취미":["축구","골프","코딩"]}
print(profile)
pickle.dump(profile, profile_file) # profile에 있는 정보를 file에 저장
profile_file.close()
```

```python
profile_file = open("profile.pickle", "rb")
profile = pickle.load(profile_file)  # 파일에 있는 것을 데이터 형태로 가져와줌
print(profile)
profile_file.close()
```
